###  Funzione lineare $f(x)=2x+1$ con $x \in [-1,1]$

A fisso $\sigma$, se il numero di dati per l'allenamento $N_{train}$ è basso, la rete ha bisogno di un numero di iterazioni $N_{epochs}$ più alto per poter imparare. Viceversa, se aumento $N_{train}$ allora posso permettermi di abbassare $N_{epochs}$, perché osservando le perdite mi accorgo che dopo un cerco valore di $N_{epochs}$ si stabilizzano e non diminuiscono più.

Per esempio, ho provato con:

- $\sigma=0.2$ e $N_{train}=10^2$, allora $N_{epochs}>100$ per avere $loss\simeq val_{loss}\simeq 0.03$.
- $\sigma=0.2$ e $N_{train}=10^3$, mi basta $N_{epochs}>20$ per avere $loss\simeq val_{loss}\simeq 0.03$.

Se invece, con queste stesse coppie di valori di $N_{train}$ e $N_{epochs}$ provo a variare $\sigma$, mi accordo che all'aumentare dell'errore la situazione precedente viene esasperata. Se per esempio, $\sigma=0.8$, allora ho necessariamente bisogno di tenere $N_{train}$ alto, altrimenti la rete pur avendo una apparente buona precisione non impara la "vera" funzione bersaglio (*underfitting*).

Se al contrario diminuisco $\sigma$, per esempio $\sigma=0.05$, mi servono sempre le stesse "proporzioni" di prima (caso $\sigma=0.2$) tra $N_{train}$ e $N_{epochs}$, ma calano le perdite e si riesce ad ottenere $loss\simeq val_{loss}\simeq 0.02$.   

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

#### Genero i dati per l'allenamento e la validazione della rete neurale: lungo l'asse $y$ "sporco" con un rumore gaussiano a media nulla e scorrelato a punti diversi.

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 100)
x_valid = np.random.uniform(-1, 1, 10)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.05 # noise standard deviation
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

#### Mostro la funzione bersaglio e il set di dati per la validazione

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=[8, 5]
plt.plot(x_valid, y_target, label='Target function')
plt.scatter(x_valid, y_valid, color='r', label='Validation data')
plt.legend(loc='upper left')
plt.grid(True); plt.show()
```

#### Genero la rete neurale scegliendo il numero e il tipo di *layer* con un rispettivo numero di neuroni (l'ultimo *layer* deve avere un solo neurone). Scelgo poi le funzioni di ottimizzazione, di perdita e la metrica. 

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,))) # layer con 1 neurone (denso) e 1 dato di input

model.compile(optimizer='sgd', loss='mse', metrics=['mse']) 
# ottimizzatore: stocastic gradient descent
# losses: mean squared error
# metrica:  mean squared error
```

```python
# get a summary of our composed model
model.summary()
```

#### Alleno la rete neurale: prende in entrata, per ogni iterazione, i dati di allenamento che ho generato prima e cerca i parametri che meglio descrivono la funzione dalla quale sono stati generati, minimizzando la funzione perdita.  

#### Stampo anche i risultati della validazione, cioè del test che faccio sul mio modello (con i dati di validazione generati in precedenza) per vedere quanto è buono il risultato. 

```python
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=250,
          shuffle=True, 
          validation_data=(x_valid, y_valid))
```

```python
# return weights
model.get_weights()
```

#### Testo il modello finale che ho generato sui dati di validazione e sulla curva esatta per vedere quanto è buono

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

#### Riporto l'andamento delle perdite in relazione alle iterazioni per l'allenamento e il test sui dati di validazione

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

#### Ora sono pronto per far predire dati sulla base del modello che ho creato e li confronto con la curva esatta

```python
x_predicted = np.random.uniform(-1, 1, 50)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r', label='Predizione')
plt.plot(x_valid, y_target, label='curva esatta')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
