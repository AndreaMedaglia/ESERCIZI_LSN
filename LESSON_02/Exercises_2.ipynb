### Random Walk in 3-Dimensioni sul reticolo

Simulazione di $M=10000$ Random Walks sul reticolo cubico con spaziatura costante $a=1$.Il punto di partenza è l'origine, ad ogni tempo discreto viene compiuto un passo di lunghezza $a$ nella direzione positiva o negativa (probabilità $1/2$ ciascuna) di uno dei 3 assi (probabilità $1/3$ ciascuno).

Grafico di $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ e della sua incertezza statistica in funzione del numero di step $i\in[0,100]$. Il risultato è fittato da una funzione $f(N)=k\sqrt{N}$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

z, g, error = np.loadtxt("./Exercise_2.2/Output/reticolo.txt", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(z,g,yerr=error, color='r', label='RW')

# FIT
def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

p_opt, p_cov = curve_fit(f, xdata=z, ydata=g)

y_fit=f(z, p_opt[0])

plt.plot(z, y_fit,'k--', label='fit')
plt.xlim([0,100])
plt.ylim([0,10])

print("optimized parameters [k] =") 
print(p_opt)
print()
print("parameters uncertainty =") 
print(np.sqrt(np.diagonal(p_cov)))

plt.legend(loc='lower right')
plt.xlabel('Step')
plt.ylabel('$r$')
plt.grid(True)
plt.show()
```

### Random Walk in 3-Dimensioni nel continuo

Simulazione di $M=10000$ Random Walks nel continuo. Il punto di partenza è l'origine, ad ogni tempo discreto viene compiuto un passo di lunghezza $a=1$ lungo una direzione casuale ottenuta campionando uniformemente un angolo solido (combinazione di $\theta \in [0,\pi]$ e $\phi \in [0,2\pi]$).

Grafico di $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ e della sua incertezza statistica in funzione del numero di step $i\in[0,100]$. Il risultato è fittato da una funzione $f(N)=k\sqrt{N}$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

z, g, error = np.loadtxt("./Exercise_2.2/Output/continuo.txt", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(z,g,yerr=error,label='RW', color='r')

# FIT
def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

p_opt, p_cov = curve_fit(f, xdata=z, ydata=g)

y_fit=f(z, p_opt[0])

plt.plot(z, y_fit, 'k--', label='fit')
plt.xlim([0,100])
plt.ylim([0,10])

print("optimized parameters [k] =") 
print(p_opt)
print()
print("parameters uncertainty =") 
print(np.sqrt(np.diagonal(p_cov)))

plt.legend(loc='lower right')
plt.xlabel('Step')
plt.ylabel('$r$')
plt.grid(True)
plt.show()
```
