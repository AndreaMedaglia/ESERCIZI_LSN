## 02.2.1 3-Dimensional Random Walk on a cubic lattice 

Simulation of $N=10000$ Random Walks on a cubic lattice with lattice constant $a=1$. At each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$. 

Picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ with its statistical uncertainty, as function of the step $i\in [0,100]$. Fit of the results with a function $f(N)=k\sqrt{N}$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

z, g, error = np.loadtxt("./Exercise_2.2.1/Output.txt", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(z,g,yerr=error)

# FIT
def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)

p_opt, p_cov = curve_fit(f, xdata=z, ydata=g)

y_fit=f(x, p_opt[0])

plt.plot(x, y_fit, 'r-', label='fit')
plt.xlim([0,100])
plt.ylim([0,10])

print("optimized parameters [k] =") 
print(p_opt)
print()
print("parameters uncertainty =") 
print(np.sqrt(np.diagonal(p_cov)))

plt.xlabel('Step')
plt.ylabel('$r$')
plt.grid(True)
plt.show()
```
