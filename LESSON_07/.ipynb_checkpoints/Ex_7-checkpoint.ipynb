## Autocorrelation function of the Potential Energy 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

t1, en = np.loadtxt("./Exercise_7/Output/Potential_energy_inst.dat", usecols=(0, 1), delimiter=';', unpack='true')

tmax=150
tplot=np.arange(0,tmax)
sum4=0
sum5=0
f=np.zeros(tmax)
den2=0

for t3 in range (0,10000):
    sum4+=en[t3]*en[t3]
    sum5+=en[t3]
    
den2=sum4/10000 - sum5**2/10000**2

for t in range(0,tmax):
    sum1=0
    sum2=0
    sum3=0
    den=0
    den=1/(10000-t)
    for t2 in range (0, (10000-t)):
        sum1+=en[t2]*en[t2+t]
        sum2+=en[t2]
        sum3+=en[t2+t]
    f[t]=(den*sum1 - den*sum2*den*sum3)/den2

plt.plot(tplot, f)
    
# FIT
def g(x,tau):  # defining the fitting function
    return np.exp(-x/tau)

p_opt, p_cov = curve_fit(g, xdata=tplot, ydata=f)

y_fit=g(tplot, p_opt[0])

plt.plot(tplot, y_fit)

print("optimized parameters [tau] =") 
print(p_opt)
print()
print("parameters uncertainty =") 
print(np.sqrt(np.diagonal(p_cov)))

plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()
```

## Statistical uncertainties as a function of blocks' size for the Potential Energy 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

en = np.loadtxt("./Exercise_7/Output/Potential_energy_inst.dat", usecols=(1), delimiter=';', unpack='true')

M=10000
L=10  # initial data for each block 
nd=500
error=np.zeros(nd)
size=np.zeros(nd)
for l in range (0,nd):
    size[l]=L
    N=M//L  # number of blocks 
    ave = np.zeros(N)
    ave2=np.zeros(N)
    sp=np.zeros(N)
    sp2=np.zeros(N)
    ep=np.zeros(N)
    

    for i in range (0,N):
        s=0
        for j in range (0,L):
            k=j+i*L
            s+=en[k]
        ave[i]=s/L
        ave2[i]=ave[i]*ave[i]

    for i in range (0,N):
        for j in range (0,i+1):
            sp[i]+=ave[j]
            sp2[i]+=ave2[j]
        sp[i]/=(i+1)
        sp2[i]/=(i+1)
        if i==0:
            ep[i]=0
        else:
            ep[i]=np.sqrt((sp2[i]-sp[i]**2)/i)

    error[l]=ep[N-1]
    L+=10

        
                
plt.plot(size,error)
plt.xscale('log')        
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()
```
