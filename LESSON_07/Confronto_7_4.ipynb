## Distribuzione radiale $g(r)$

Confronto tra le distribuzioni radiali $g(r)$ (nelle tre fasi) calcolate tramite algoritmo di dinamica molecolare *MD* (lezione 04) e metodi Montecarlo *MC* (lezione 07).   

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

r_l, g_l, error_l = np.loadtxt("./Exercise_7/Output/output.gave_l.0", usecols=(0,2,3), unpack='true')
plt.errorbar(r_l,g_l,yerr=error_l, label='liquid MC')
x_l, f_l, err_l = np.loadtxt("./Output_4/blocking.gave_l.dat", usecols=(0,2,3), unpack='true')
plt.errorbar(x_l,f_l,yerr=err_l, label='liquid MD')
plt.title('radial distribution g(r) for the liquid phase') 
plt.xlabel('r')
plt.ylabel('g(r)')
plt.legend(loc='upper right')
plt.grid(True)


plt.figure()
r_l, g_l, error_l = np.loadtxt("./Exercise_7/Output/output.gave_s.0", usecols=(0,2,3), unpack='true')
plt.errorbar(r_l,g_l,yerr=error_l, label='solid MC')
x_l, f_l, err_l = np.loadtxt("./Output_4/blocking.gave_s.dat", usecols=(0,2,3), unpack='true')
plt.errorbar(x_l,f_l,yerr=err_l, label='solid MD')
plt.title('radial distribution g(r) for the solid phase') 
plt.xlabel('r')
plt.ylabel('g(r)')
plt.xlim()
plt.legend(loc='upper right')
plt.grid(True)

plt.figure()
r_l, g_l, error_l = np.loadtxt("./Exercise_7/Output/output.gave_g.0", usecols=(0,2,3), unpack='true')
plt.errorbar(r_l,g_l,yerr=error_l, label='gas MC')
x_l, f_l, err_l = np.loadtxt("./Output_4/blocking.gave_g.dat", usecols=(0,2,3), unpack='true')
plt.errorbar(x_l,f_l,yerr=err_l, label='gas MD')
plt.title('radial distribution g(r) for the gas phase') 
plt.xlabel('r')
plt.ylabel('g(r)')
plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```

## Energia potenziale per particella $U/N$

Confronto tra le energie potenziali per particella $U/N$ (nelle tre fasi) calcolate tramite algoritmo di dinamica molecolare *MD* (lezione 04) e metodi Montecarlo *MC* (lezione 07). I risultati sono riportati in unit di Lennard-Jones.  

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("./Exercise_7/Output/output.epot_g.0", usecols=(0,2,3), unpack='true')
y, g, err = np.loadtxt("./Output_4/blocking.epot_g.dat", usecols=(0,2,3), unpack='true')
plt.errorbar(x,f,yerr=error, label='MC')
plt.errorbar(y,g,yerr=err, label='MD')
plt.title('U/N gas phase') 
plt.xlabel('block')
plt.ylabel('U(N,T)')
plt.legend(loc='best')
plt.grid(True)

plt.figure()
x, f, error = np.loadtxt("./Exercise_7/Output/output.epot_l.0", usecols=(0,2,3), unpack='true')
y, g, err = np.loadtxt("./Output_4/blocking.epot_l.dat", usecols=(0,2,3), unpack='true')
plt.errorbar(x,f,yerr=error, label='MC')
plt.errorbar(y,g,yerr=err, label='MD')
plt.title('U/N liquid phase') 
plt.xlabel('block')
plt.ylabel('U(N,T)')
plt.legend(loc='best')
plt.grid(True)

plt.figure()
x, f, error = np.loadtxt("./Exercise_7/Output/output.epot_s.0", usecols=(0,2,3), unpack='true')
y, g, err = np.loadtxt("./Output_4/blocking.epot_s.dat", usecols=(0,2,3), unpack='true')
plt.errorbar(x,f,yerr=error, label='MC')
plt.errorbar(y,g,yerr=err, label='MD')
plt.title('U/N solid phase') 
plt.xlabel('block')
plt.ylabel('U(N,T)')
plt.legend(loc='best')
plt.grid(True)


plt.show()
```
