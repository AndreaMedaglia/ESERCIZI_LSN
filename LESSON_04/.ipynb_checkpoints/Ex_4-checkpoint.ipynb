## TEMPERATURE 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("./Ex_4.1/Output/blocking_temp.dat", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('Blocks')
plt.ylabel('T')
plt.grid(True)
plt.show()
```
