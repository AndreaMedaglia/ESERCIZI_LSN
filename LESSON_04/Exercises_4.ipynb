## ESERCIZIO 04.1

Il programma legge dal file *input.dat* i parametri per la simulazione e dal file *config.0* la configurazione attuale al tempo $t$. Per usare l'algoritmo di Verlet, però, mi serve anche una configurazione "vecchia", al tempo $t-\delta t$. Il programma quindi chiede da terminale:

- se si vuole generare la configurazione "vecchia" a partire da velocità generate casualmente e riscalate opportunamente in base alla temperatura di input.
- se si vuole leggere la configurazione "vecchia" dal file *old.0* (se il file non esiste, il programma esegue il punto precedente).

Se la configurazione "vecchia" è letta dal file *old.0*, il programma chiede, sempre su terminale, se si vuole riscalare tale configurazione in modo da avere la temperatura desiderata. 

Successivamente, a prescindere dalla scelta effettuata in precedenza, il programma chiede se si vuole equilibrare il sistema:
- Se la risposta è no, il programma entra nella simulazione "vera e propria". 
- Se è sì, bisogna inserire su terminale un intero $n$, che rappresenta i passi Verlet di cui viene simulato il sistema. Alla fine di questa simulazione, il programma calcola la temperatura in base alle configurazioni ottenute, la stampa su terminale e poi chiede se si vuole riscalare. A questo punto la situazione è la stessa la stessa in cui ci si trova quando il programma chiede se si vuole equilibrare il sistema. 

Questa procedura (ripetibile all'infinito) è da effettuare finchè la temperatura calcolata a partire dalle configurazioni ottenute non è simile a quella di simulazione (letta da file *input.dat*).

### Calcolo della funzione di distribuzione radiale $g(r)$

Calcolo della funzione di distribuzione radiale $g(r)$ in funzione della distanza radiale $r \in \left[ 0; L/2\right]$. Nel grafico sono rappresentate contemporaneamente le tre fasi e i valori medi finali con incertezza.   

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

r_l, g_l, error_l = np.loadtxt("./Exercise_4.1/Output/blocking.gave_l.dat", usecols=(0,2,3), unpack='true')
r_s, g_s, error_s = np.loadtxt("./Exercise_4.1/Output/blocking.gave_s.dat", usecols=(0,2,3), unpack='true')
r_g, g_g, error_g = np.loadtxt("./Exercise_4.1/Output/blocking.gave_g.dat", usecols=(0,2,3), unpack='true')
plt.errorbar(r_l,g_l,yerr=error_l, label='liquid phase')
plt.errorbar(r_s,g_s,yerr=error_s, label='solid phase')
plt.errorbar(r_g,g_g,yerr=error_g, label='gas phase')


plt.title('radial distribution g(r) for the three phases') 
plt.xlabel('r')
plt.ylabel('g(r)')
plt.xlim(left=0)
plt.xlim(right=2.3)
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
