### Problema del commesso viaggiatore con città uniformemente distribuite nel quadrato $[0,1]\times [0,1]$ 

Vengono generate $30$ città nel quadrato di lato unitario e con un *simulated annealing* si ricerca il percorso ottimale che passi una e una sola volta per ogni città tornando al punto di partenza. La temperatura iniziale è $T=1/\beta$ con $\beta=1$, quella finale ha $\beta=200$, gli incrementi sono di $0.1$. Il numero di iterazioni, quindi, è $N_{iter}=1991$. Ad ogni iterazione vengono compiuti $10^3$ passi Metropolis. Con lunghezza, da qui in avanti, intendo $L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N)$.

Ogni passo Metropolis consiste nell'effettuare o una permutazione o uno *shift*, con $prob=50\%$ a testa. 

Il primo grafico rappresenta il percorso alla temperatura finale. Per evidenziare il punto di partenza e di arrivo è lasciata "aperta" la linea che unisce le città ma ovviamente ai fini del calcolo della lunghezza del percorso, nel codice, viene considerata anche la strada per tornare al punto iniziale.

Il secondo grafico rappresenta la lunghezza del percorso finale di ogni iterazione (cioè ad ogni temperatura).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize']=[8, 5]

x, y = np.loadtxt("./Exercise_10.1/Output/Configuration_s.dat", usecols=(0,1), unpack='true')
plt.plot(x, y, 'b')
plt.plot(x, y, 'o', color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Percorso finale')
plt.grid(True)

plt.figure()
x, y = np.loadtxt("./Exercise_10.1/Output/Best_l_s.dat", usecols=(0,1), unpack='true')
plt.plot(x, y)
plt.xlabel('N_iter')
plt.ylabel('length')
plt.title('Lunghezza del percorso finale per ogni iterazione')
plt.grid(True)

print()
print('Lunghezza del miglior percorso = ', y[-1])
print()

plt.show()
```

### Problema del commesso viaggiatore con città uniformemente distribuite sulla circonferenza unitaria

Vengono generate $30$ città sulla circonferenza unitaria e con un *simulated annealing* si ricerca il percorso ottimale che passi una e una sola volta per ogni città tornando al punto di partenza. La temperatura iniziale è $T=1/\beta$ con $\beta=1$, quella finale ha $\beta=50$, gli incrementi sono di $0.1$. Il numero di iterazioni, quindi, è $N_{iter}=491$. Ad ogni iterazione vengono compiuti $10^3$ passi Metropolis. Con lunghezza, da qui in avanti, intendo $L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N)$.

$\beta$ finale è più piccolo rispetto al caso precedente perché per l'algoritmo è più "facile" trovare il minimo. Quindi ai fini di dispendio computazionale si può tenere più basso senza perdere precisione sul risultato finale. 

Ogni passo Metropolis consiste nell'effettuare o una permutazione o uno *shift*, con $prob=50\%$ a testa. 

Il primo grafico rappresenta il percorso alla temperatura finale. Per evidenziare il punto di partenza e di arrivo è lasciata "aperta" la linea che unisce le città ma ovviamente ai fini del calcolo della lunghezza del percorso, nel codice, viene considerata anche la strada per tornare al punto iniziale.

Il secondo grafico rappresenta la lunghezza del percorso finale di ogni iterazione (cioè ad ogni temperatura).

```python
x, y = np.loadtxt("./Exercise_10.1/Output/Configuration_c.dat", usecols=(0,1), unpack='true')
plt.plot(x, y, 'b')
plt.plot(x, y, 'o', color='r')
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Miglior percorso')
plt.grid(True)

plt.figure()
x, y = np.loadtxt("./Exercise_10.1/Output/Best_l_c.dat", usecols=(0,1), unpack='true')
plt.plot(x, y)
plt.xlabel('$N_{iter}$')
plt.ylabel('length')
plt.title('Lunghezza del miglior percorso per ogni iterazione')
plt.grid(True)

print()
print('Lunghezza del miglior percorso = ', y[-1])
print()

plt.show()
```
