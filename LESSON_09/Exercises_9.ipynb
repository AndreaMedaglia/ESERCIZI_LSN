### Problema del commesso viaggiatore con città uniformemente distribuite nel quadrato $[0,1]\times [0,1]$ 

Vengono generate $30$ città nel quadrato di lato unitario e con un algoritmo genetico si ricerca il percorso ottimale che passi una e una sola volta per ogni città tornando al punto di partenza. Il numero di *cromosomi* per popolazione è $10^3$, il numero di generazioni è $5\times 10^2$. Con lunghezza, da qui in avanti, intendo $L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N)$.

I seguenti grafici sono ottenuti con $prob=50\%$ di crossover, $prob=10\%$ di permutazioni, $prob=7\%$ di effettuare uno *shift* e $prob=5\%$ di effettuare una inversione a specchio della lista di città.

Il primo grafico rappresenta il miglior percorso. Per evidenziare il punto di partenza e di arrivo è lasciata "aperta" la linea che unisce le città ma ovviamente ai fini del calcolo della lunghezza del percorso, nel codice, viene considerata anche la strada per tornare al punto iniziale.

Il secondo grafico rappresenta la lunghezza del miglior percorso, cioè il primo nella popolazione ordinata, per ogni generazione.

Il terzo grafico invece rappresenta la lunghezza mediata sui primi $\frac{10^3}{2}$ elementi, cioè i migliori, della popolazione ordinata.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize']=[8, 5]

x, y = np.loadtxt("./Exercise_9/Output/Configuration_s.dat", usecols=(0,1), unpack='true')
plt.plot(x, y, 'b')
plt.plot(x, y, 'o', color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Miglior percorso')
plt.grid(True)

plt.figure()
x, y = np.loadtxt("./Exercise_9/Output/Best_l_s.dat", usecols=(0,1), unpack='true')
plt.plot(x, y)
plt.xlabel('N_iter')
plt.ylabel('length')
plt.title('Lunghezza relativa al miglior percorso per ogni generazione')
plt.grid(True)

print()
print('Lunghezza del miglior percorso = ', y[-1])
print()


plt.figure()
x, y = np.loadtxt("./Exercise_9/Output/Average_l_s.dat", usecols=(0,1), unpack='true')
plt.plot(x, y)
plt.xlabel('N_iter')
plt.ylabel('length')
plt.title('Media delle lunghezze (sulla miglior metà della popolazione)')
plt.grid(True)

plt.show()
```
