```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

points=100
T = np.linspace(0.5,2.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e)

x, f, error = np.loadtxt("./ISING_1D/Internal_energy.dat", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('T')
plt.ylabel('U(N,T)')
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

points=100
T = np.linspace(0.5,2.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X)

x, f, error = np.loadtxt("./ISING_1D/Susceptivity.dat", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.grid(True)
plt.show()
```
