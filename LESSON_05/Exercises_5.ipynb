## Caso uniforme

### Distribuzione dei punti campionati nel piano 3D del *ground state*

Grafico dei punti campionati usando una probabilità di transizione uniforme $T(\vec{x}|\vec{y})$ nel piano $xyz$. Le distanze sono espresse in unità del raggio di Bohr $a_0$; in numero di punti campionati è $10^5$.

Lo spostamento di ogni passo di Metropolis è stato implementato con un *Random Walk* tridimensionale: con probabilità 1/3 viene scelto un asse e con probabilità 1/2 il senso positivo o negativo dello stesso. Il punto di partenza è l'origine; gli spostamenti sono generati casualmente e uniformemente nell'intervallo $[0.75, 1.55]$, numeri opportuni per avere circa il $50\%$ di accettanza.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z = np.loadtxt("./Exercise_5.1/Output/Coordinate_gs_unif.txt", usecols=(0,1,2), delimiter=';', unpack='true')

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

### Distanza media tra i punti campionati nel piano 3D del *ground state* 

Grafico della stima della distanza media $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ (in unità del raggio di Bohr) e della sua incertezza (calcolate con la tecnica del datablocking) in funzione del numero di blocchi $N=100$. 

L'algoritmo di Metropolis è implementato come descritto sopra.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("./Exercise_5.1/Output/Distanza_gs_unif.txt", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('Blocks')
plt.ylabel('<r>/$a_0$')
plt.grid(True)
plt.show()
```
