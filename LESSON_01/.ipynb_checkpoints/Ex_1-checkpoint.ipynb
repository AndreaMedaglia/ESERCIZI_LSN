## ESERCIZIO 01.2

### Estensione del generatore di numeri pseudo-causali per una generica distribuzione esponenziale e prova del Teorema del Limite Centrale 

Estrazione di numeri pseudo-casuali $\{x_i\}_i$ da una distribuzione $p(x) = \lambda \exp(-\lambda x)$ con $x\in [0;+\infty]$ e $\lambda=1$. 

Gli istogrammi sottostanti sono ottenuti riempiendo i *bins* con $N=10^4$ realizzazioni di $S_L = \frac{1}{L}\sum_{i=1}^L x_i$ con $L=1, 2, 10, 100$, e di conseguenza generando in totale $M=N \times L$ numeri casuali.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

n_bins = 100

x1 = np.loadtxt("./Exercise_1.2/Output/exp_1.txt", usecols=(0), unpack='true')
x2 = np.loadtxt("./Exercise_1.2/Output/exp_2.txt", usecols=(0), unpack='true')
x10 = np.loadtxt("./Exercise_1.2/Output/exp_10.txt", usecols=(0), unpack='true')
x100 = np.loadtxt("./Exercise_1.2/Output/exp_100.txt", usecols=(0), unpack='true')

plt.hist(x1, n_bins, range=(0,4), label='L=1')
plt.hist(x2, n_bins, range=(0,4), label='L=2')
plt.hist(x10, n_bins, range=(0,4), label='L=10')
plt.hist(x100, n_bins, range=(0,4), label='L=100')

plt.xlabel('S_L')
plt.ylabel('\N{greek small letter nu}')
plt.title('Exponential dice')

plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

### Estensione del generatore di numeri pseudo-causali per una generica distribuzione lorentziana e prova del Teorema del Limite Centrale

Estrazione di numeri pseudo-casuali $\{x_i\}_i$ da una distribuzione $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$ con $x\in [-\infty;+\infty]$, $\Gamma=1$ e $\mu=0$.

Gli istogrammi sottostanti sono ottenuti riempiendo i *bins* con $N=10^4$ realizzazioni di $S_L = \frac{1}{L}\sum_{i=1}^L x_i$ con $L=1, 2, 10, 100$, e di conseguenza generando in totale $M=N \times L$ numeri casuali.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

n_bins = 100

x1 = np.loadtxt("./Exercise_1.2/Output/cl_1.txt", usecols=(0), unpack='true')
x2 = np.loadtxt("./Exercise_1.2/Output/cl_2.txt", usecols=(0), unpack='true')
x10 = np.loadtxt("./Exercise_1.2/Output/cl_10.txt", usecols=(0), unpack='true')
x100 = np.loadtxt("./Exercise_1.2/Output/cl_100.txt", usecols=(0), unpack='true')

plt.hist(x1, n_bins, range=(-10,10), label='L=1')
plt.hist(x2, n_bins, range=(-10,10), label='L=2')
plt.hist(x10, n_bins, range=(-10,10), label='L=10')
plt.hist(x100, n_bins, range=(-10,10), label='L=100')

plt.xlabel('S_L')
plt.ylabel('\N{greek small letter nu}')
plt.title('Lorentzian dice')

plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

### Prova del Teorema del Limite Centrale con numeri pseudo-casuali uniformemente generati 

Estrazione di numeri pseudo-casuali $\{x_i\}_i$ uniformemente distribuiti in $x\in [0;1)$.

Gli istogrammi sottostanti sono ottenuti riempiendo i *bins* con $N=10^4$ realizzazioni di $S_L = \frac{1}{L}\sum_{i=1}^L x_i$ con $L=1, 2, 10, 100$, e di conseguenza generando in totale $M=N \times L$ numeri casuali.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

n_bins = 100

x1 = np.loadtxt("./Exercise_1.2/Output/u_1.txt", usecols=(0), unpack='true')
x2 = np.loadtxt("./Exercise_1.2/Output/u_2.txt", usecols=(0), unpack='true')
x10 = np.loadtxt("./Exercise_1.2/Output/u_10.txt", usecols=(0), unpack='true')
x100 = np.loadtxt("./Exercise_1.2/Output/u_100.txt", usecols=(0), unpack='true')

plt.hist(x1, n_bins, range=(0,1), label='L=1')
plt.hist(x2, n_bins, range=(0,1), label='L=2')
plt.hist(x10, n_bins, range=(0,1), label='L=10')
plt.hist(x100, n_bins, range=(0,1), label='L=100')

plt.xlabel('S_L')
plt.ylabel('\N{greek small letter nu}')
plt.title('Standard dice')

plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

### Fit degli istogrammi con $L=100$ dei tre precedenti casi 

#### Numeri generati da distribuzione esponenziale: fitto con una gaussiana.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

x_100 = np.loadtxt("./Exercise_1.2/Output/exp_100.txt", usecols=(0), unpack='true')

n_bins = 100

n, bins, patches = plt.hist(x_100, n_bins, range=(0.5,1.5), label='L=100')

plt.xlabel('S_L')
plt.ylabel('\N{greek small letter nu}')
plt.title('Exponential dice')
plt.legend(loc='upper right')

# FIT
def f(x,a,b,c):  # defining the fitting function
    return a * np.exp(-b*(x-c)**2)

x=np.linspace(0.5, 1.5, n_bins)
p_opt, p_cov = curve_fit(f, xdata=x, ydata=n)


y_fit=f(x, p_opt[0],p_opt[1],p_opt[2])

plt.plot(x, y_fit, 'r-', label='fit')

print("optimized parameters [a,b,c] =") 
print(p_opt)
print()
print("parameters uncertainty =") 
print(np.sqrt(np.diagonal(p_cov)))



plt.grid(True)
plt.show()

```

#### Numeri generati da distribuzione di Cauchy-Lorentz: fitto con una lorentziana.

```python
# import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

x_100 = np.loadtxt("./Exercise_1.2/Output/cl_100.txt", usecols=(0), unpack='true')

n_bins = 100

n, bins, patches = plt.hist(x_100, n_bins, range=(-10,10), label='L=100')

plt.xlabel('S_L')
plt.ylabel('\N{greek small letter nu}')
plt.title('Lorentian dice')
plt.legend(loc='upper right')

# FIT
def f(x,m,g,a):  # defining the fitting function
    return a*(g/np.pi)*(1/((x-m)**2+g**2))

x=np.linspace(-10, 10, n_bins)

p_opt, p_cov = curve_fit(f, x, n)
y_fit=f(x, p_opt[0],p_opt[1], p_opt[2])
plt.plot(x, y_fit, 'r-')

print("optimized parameters [m,gamma,a] =") 
print(p_opt)
print()
print("parameters uncertainty =") 
print(np.sqrt(np.diagonal(p_cov)))



plt.grid(True)
plt.show()
```

#### Numeri generati da distribuzione uniforme: fitto con una gaussiana.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

x_100 = np.loadtxt("./Exercise_1.2/Output/u_100.txt", usecols=(0), unpack='true')

n_bins = 100

n, bins, patches = plt.hist(x_100, n_bins, range=(0.35,0.65), label='L=100')

plt.xlabel('S_L')
plt.ylabel('\N{greek small letter nu}')
plt.title('Standard dice')
plt.legend(loc='upper right')

# FIT
def f(x,a,b,c):  # defining the fitting function
    return a * np.exp(-b*(x-c)**2)

x=np.linspace(0.35, 0.65, n_bins)
p_opt, p_cov = curve_fit(f, xdata=x, ydata=n)


y_fit=f(x, p_opt[0],p_opt[1],p_opt[2])

plt.plot(x, y_fit, 'r-', label='fit')

print("optimized parameters [a,b,c] =") 
print(p_opt)
print()
print("parameters uncertainty =") 
print(np.sqrt(np.diagonal(p_cov)))



plt.grid(True)
plt.show()

```
