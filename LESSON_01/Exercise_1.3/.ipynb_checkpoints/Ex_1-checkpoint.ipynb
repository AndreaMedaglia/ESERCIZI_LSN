## 01.1.1 Estimation of $\langle r \rangle$ and its uncertainty 


Estimation and picture of $\langle r \rangle = \int_0^1 r dr = 1/2$ with a number of *throws* $M=100000$ as a function of the number of blocks $N=100$.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("Output.txt", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(x,f-0.5,yerr=error)
plt.xlabel('Blocks')
plt.ylabel('<r> - 1/2')
plt.grid(True)
plt.show()
```

## 01.1.2 Estimation of $\sigma^2$ and its uncertainty  

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("Output2.txt", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(x,f-1/12,yerr=error)
plt.xlabel('Blocks')
plt.ylabel('<$\sigma$> - 1/12')
plt.grid(True)
plt.show()
```
