### Calcolo del prezzo della *European call-option* $C[S(0),0]$ al tempo $t=0$ campionando direttamente il prezzo finale $S(T)$

Stima della *European call-option* $C[S(0),0]=e^{-rT}\times max(0,<S>-K)$ e della sua incertezza (calcolate con la tecnica del datablocking) in funzione del numero di blocchi $N=100$, ottenuta campionando direttamente il prezzo finale $S(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$, con $W(T)\sim\mathcal{N}(0,T)$. Il numero di punti campionati è $M=10^4$. 

I parametri sono (validi anche per i successivi esercizi):
- prezzo dell'*asset* a $t=0$: $S(0)=100$
- tempo finale: $T=1$
- prezzo d'esercizio: $K=100$
- *rate* di interesse a rischio nullo: $r=0.1$
- volatilità: $\sigma=0.25$



```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("./Exercise_3.1/Output/diretto_C.txt", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(x,f,yerr=error, label='Numerica')

y=14.975790778311286
plt.axhline(y, color='r', label='B-S')

plt.xlabel('Blocks')
plt.ylabel('C')
plt.grid(True)

plt.legend()


plt.show()
```

### Calcolo del prezzo della *European call-option* $C[S(0),0]$ al tempo $t=0$ campionando $S(T)$ dal Moto Browniano Geometrico $GBM(\mu,\sigma^2)$ discretizzato 

Stima della *European call-option* $C[S(0),0]=e^{-rT}\times max(0,<S>-K)$ e della sua incertezza (calcolate con la tecnica del datablocking) in funzione del numero di blocchi $N=100$, ottenuta campionando il Moto Browniano Geometrico $GBM(\mu,\sigma^2)$ discretizzato  $S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$, con $Z_i \sim \mathcal{N}(0,1)$. Il numero di punti campionati è $M=10^4$, l'intervallo temporale $[0,T]$ è diviso in $10^2$ sottointervalli. 


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("./Exercise_3.1/Output/discreto_C.txt", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(x,f,yerr=error, label='Numerica')

y=14.975790778311286
plt.axhline(y, color='r', label='B-S')

plt.xlabel('Blocks')
plt.ylabel('C')
plt.grid(True)

plt.legend()

plt.show()
```
