# EXERCISE 03



## Direct Call

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, f, error = np.loadtxt("./Exercise_3.1/Output_C.txt", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.xlabel('Blocks')
plt.ylabel('C')
plt.grid(True)
plt.show()
```
